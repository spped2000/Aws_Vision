```python
panel = cv2.imread('panel.png')

rect_roi = roi_helper.draw_rectangle(panel, 4)  # quantity=3 specifies number of rectangles to draw

frame_temp = roi_helper.visualize_roi(panel, rect_roi)
```

```python
panel = cv2.cvtColor(panel, cv2.COLOR_BGR2RGB)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches



# Create figure and axes
fig, ax = plt.subplots(1)

# Set larger figure size
fig.set_size_inches(12, 8)

# Create each rectangle from the dict and add to the axes
for roi in rect_roi['roi'].values():
    ax.imshow(panel)
    rect = patches.Rectangle((roi['tl_x'], roi['tl_y']), roi['w'], roi['h'], linewidth=1, edgecolor='r', facecolor='none')
    ax.add_patch(rect)
   

# Set limits
plt.xlim(0, 800)
plt.ylim(0, 600)

# We invert the y axis, because in matplotlib the origin is at the top left
plt.gca().invert_yaxis()

# Show the figure with the rectangles
plt.show()
```
